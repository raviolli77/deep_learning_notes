```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from PIL import Image, ImageFilter

%matplotlib inline
```

```python
im = Image.open('mammal.jpg')
```

```python
im_alt_array = np.asarray(im)
```

```python
print(im_alt_array.shape, im_alt_array.dtype)
```

```python
plt.imshow(im_alt_array)
plt.axis('off')
plt.title('Issa Ram')
plt.show()
```

```python
im_color_array = np.asarray(im.resize((128, 128)))
```

```python
im_color_array.shape
```

```python
plt.imshow(im_color_array)
plt.axis('off')
plt.title('Issa Resized Ram')
plt.show()
```

```python
im_color_array.reshape(-1)
```

Creating an array with the same size as our image array, filling in all values with zero. Then iteratively grabbing the respective values for each RGB instance and filling in those values into the array with all zero values. Then plotting each respective outcome to receive the $R$ or $G$ or $B$ plots, where $c = 0 (red), 1(green), 2(blue)$. 

From [Iain's article](http://www.degeneratestate.org/posts/2016/Oct/23/image-processing-with-numpy/)

```python
fig, axs = plt.subplots(1, 3, figsize=(15,5))

for c, ax in zip(range(3), axs):
    tmp_im = np.zeros(im_alt_array.shape, dtype="uint8")
    tmp_im[:,:,c] = im_alt_array[:,:,c]
    ax.imshow(tmp_im)
    ax.set_axis_off()
```

Purple for shits and giggles

```python
tmp_im2 = np.zeros(im_alt_array.shape, 
                   dtype="uint8")
```

```python
tmp_im2[:,:,0] = im_alt_array[:,:,0] # Filling in array with Red values for image
tmp_im2[:,:,2] = im_alt_array[:,:,2] # Filling in array with Blue values for image
# Red + Blue = Purple
```

```python
plt.imshow(tmp_im2)
plt.axis('off')
plt.title('Issa Purple Goat')
plt.show()
```

# GreyScale Manipulation

```python
# Greyscale
im_alt = im.convert('L')
```

```python
im_array = np.asarray(im_alt, 'f')
```

```python
plt.imshow(im_array,
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Issa Grey Ram')
plt.show()
```

```python
print(im_array.shape, im_array.dtype)
```

```python
# Invert the image
plt.imshow(255 - im_array,
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Issa Inverted Greyscale Ram')
plt.show()
```

```python
plt.imshow(im_array[:685, :],
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Top Half of the Ram')
plt.show()
```

```python
plt.imshow(im_array[685:, :],
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Bottom Half of the Ram')
plt.show()
```

```python
plt.imshow(im_array[:, :960],
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Left Half of the Ram')
plt.show()
```

```python
plt.imshow(im_array[:, 960:],
          cmap = plt.get_cmap('gray'))
plt.axis('off')
plt.title('Right Half of the Ram')
plt.show()
```
